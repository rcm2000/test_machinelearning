```python
import sklearn
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
nyctaxi = pd.read_csv('./files/nyc-taxi-trip-duration_train.csv')
```

```python
nyctaxi.info()
```

```python
nyctaxi
```

```python
nyctaxi['x'] = (nyctaxi.pickup_longitude - nyctaxi.dropoff_longitude)**2
nyctaxi['y'] = (nyctaxi.pickup_latitude - nyctaxi.dropoff_latitude)**2
nyctaxi['distance_sqrt'] = nyctaxi.x + nyctaxi.y
nyctaxi['distance'] = np.sqrt(nyctaxi.distance_sqrt) * 1000
nyctaxi['speed'] = nyctaxi.distance / nyctaxi.trip_duration * 1000
```

```python
nyctaxi
```

```python
x = nyctaxi[['distance']]
y = nyctaxi[['trip_duration']]
```

```python
x.shape , y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x,y)
X_train.shape ,X_test.shape ,Y_train.shape ,Y_test.shape 
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x,y)
```

```python
lr.coef_,lr.intercept_
```

```python
lr.score(x,y)
```

```python
lr.score(X_train,Y_train)
```
